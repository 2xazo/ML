```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df=pd.read_csv('sales.csv')  #6
```

```python
df
```

```python
df.columns
```

```python
from sklearn.linear_model import LinearRegression
x=df[['Weeks']]
y=df['Total']
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y,y_pred)
mse=mean_squared_error(y,y_pred)
print("mena sq ",mse)
print("mena abs ",MAE)
```

```python
df['predation']=model.predict(x)
```

```python
df
```

```python
xpoint=df['Weeks']
ypoint=df['Total']
yprdationpoint=df['predation']
plt.scatter(xpoint,ypoint)
plt.plot(xpoint,yprdationpoint,color="red")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=False)
xnew=poly.fit_transform(x)
print(xnew)
```

```python
model2=LinearRegression()
model2.fit(xnew,y)
```

```python
y_pred2=model2.predict(xnew)
MAE=mean_absolute_error(y,y_pred2)
mse=mean_squared_error(y,y_pred2)
print("mena sq ",mse)
print("mena abs ",MAE)
```

```python
df['non-liner predacation']=model2.predict(xnew)
xpoint=df['Weeks']
ypoint=df['Total']
yprdationpoint=df['non-liner predacation']
plt.scatter(xpoint,ypoint)
plt.plot(xpoint,yprdationpoint,color="red")
plt.show()
```
